### PROBABILITY DISTRIBUTIONS 

 Class Example:
 The probability distribution for the personality inventory test for passive- aggressive traits discussed in class is given below. Find the mean score. 
 
 x | 1 |2|3|4|5|
 --|---|-|-|-|-|
 p(x)|0.16 |0.22 |0.28 |0.20 |0.14|

```python
#define values
values = [1, 2, 3, 4, 5]

#define probabilities
probs  = [0.16, 0.22, 0.28, 0.20, 0.14]

values = np.asarray(values)
probs = np.asarray(probs)
```

```python
mean_value = (values * probs).sum() / probs.sum()
print('%.2f' % mean_value)
```

```python
variance = (pow(values-mean_value,2)*probs).sum()
print('%.3f' % variance)
```

```python
std_dev = math.sqrt(variance)
print('%.2f' % std_dev)
```

# Practice: Try to do it yourself!!

Students in a class take a quiz with eight questions. The random variable $x$ represents the number of questions answered correctly.

x | 0|1|2|3 |4| 5| 6 |7 |8|
--|--|-|-|--|-|--|---|--|-|
P(x)| 0.02|0.02 |0.06|0.06 |0.08 |0.22 |0.30|0.16 |0.08

Use the probability distribution or histogram to find the:

* mean,
* variance, 
* standard deviation, 
* expected value of the probability distribution, and
* interpret the results. 

```python
value = [0,1,2,3,4,5,6,7,8]
probability = [0.02, 0.02, 0.06, 0.06, 0.08, 0.22, 0.30, 0.16, 0.08]
value = np.asarray(value)
probability = np.asarray(probability)
plt.bar(value, probability)
plt.title("Correct responses by Students in a Quiz", color="black")
```

```python
mean_value = (value * probability).sum() / probability.sum()
print('%.2f' % mean_value)
```

```python
variance = (pow(value-mean_value,2)*probability).sum()
print('%.2f' % variance)
```

```python
std_dev = math.sqrt(variance)
print('%.2f' % std_dev)
```

```python
expected_value = (value * probability).sum() / probability.sum()
print('%.2f' % expected_value)
```

```python
#Interpretation 
# On an average, approximately 5 questions are answered correctly by students. 
# 30% of the students have answered 6 questions correctly.
# Through the histogram, it is evident that the distribution is left skewed.
```
