```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
```

```python
dataset= pd.read_csv('/Users/devi/Documents/Devi/MPS Analytics/Probability and Statistics/Module 5/AirQualityUCI.csv')
#print(dataset)
dataset.head()
```

```python
t=dataset['T']
rh=dataset['RH']
ah=dataset['AH']
cogt=dataset['CO(GT)']
c6h6gt=dataset['C6H6(GT)']
```

```python
# select the 5 columns
selected_cols = ['T', 'RH' , 'AH' , 'CO(GT)' , 'C6H6(GT)']
dataset_selected = dataset[selected_cols]
sns.heatmap(dataset_selected.corr())
print(dataset_selected.corr())
```

```python
#Correlation between Temperature and Relative Humidity
plt.scatter(t,rh,color = 'purple')
plt.title("Temperature Vs Relative Humidity")
plt.ylabel('Temperature')
plt.xlabel('Relative Humidity')
plt.show()
r, p = pearsonr(t, rh)
print('Pearsons correlation: %.3f' % r)
print('P-value: %.4f' % p)
```
